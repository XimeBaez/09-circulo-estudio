```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('tmp/Proyectos.csv')

df.head()
```

### Distribicion de proyectos por año

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Ejercicio', data=df, palette='coolwarm')
plt.title('Distribucion de proyectos por anio')
plt.xlabel('Año')
plt.ylabel('Cantidad de proyectos')

plt.show()
```

### Duracion promedio de los proyectos por rubro

```python
df.head()
```

```python
df['Fecha de Inicio'] = pd.to_datetime(df['Fecha de Inicio'])
df['Fecha de Finalizacion'] = pd.to_datetime(df['Fecha de Finalizacion'])

df['Duracion (Dias)'] = (df['Fecha de Finalizacion'] - df['Fecha de Inicio']).dt.days

df.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Rubro', y='Duracion (Dias)', data=df, estimator=lambda x: x.mean(), palette='magma')
plt.title('Duracion promedio de proyectos')
plt.xlabel('Rubro')
plt.ylabel('Duracion en dias')
plt.xticks(rotation=45)

plt.show()
```

### Mapa de calor por tipo de recurso y rubro

```python
pivot_table = df.pivot_table(
    index='Tipo de Recurso',
    columns='Rubro',
    aggfunc='size',
    fill_value=0
)


plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, annot=True, cmap='Blues')

plt.title('Proyectos por tipo de recurso y rubro')
plt.xlabel('Rubro')
plt.ylabel('Tipo de recurso')

plt.show()
```

### Proyectos comunes por institucion 

```python
df.head()
```

```python
df['Instituciones'] = df['Instituciones participantes'].str.split('.')
df.head()
```

```python
instituciones = df.explode('Instituciones')['Instituciones'].value_counts()
instituciones
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=instituciones.values[:10], y=instituciones.index[:10], palette='Set2')

plt.title('Instituciones que mas participan ')
plt.xlabel('Cantidad de participantes')
plt.ylabel('Instituciones')

plt.show()
```

### Proyectos por convocatoria

```python
plt.figure(figsize=(10,6))
df['Convocatoria'].value_counts()[:10].plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='coolwarm')

plt.title('Proyectos por convocatoria')
plt.ylabel('')
plt.show()
```
